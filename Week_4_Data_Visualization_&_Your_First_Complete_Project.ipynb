1️⃣ DATA LOADING

```python
import pandas as pd
import matplotlib.pyplot as plt\

# Sales Data
sales_data = pd.DataFrame({
    "Category": ["Electronics", "Clothing", "Groceries", "Toys", "Furniture"],
    "Sales": [45000, 30000, 50000, 15000, 28000]
})

# Temperature Data
temperature_data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Temperature": [18, 20, 24, 28, 32, 34, 33, 32, 29, 25, 22, 19]
})

# Expense Data
expense_data = pd.DataFrame({
    "Category": ["Rent", "Food", "Transport", "Entertainment", "Utilities"],
    "Amount": [15000, 8000, 3000, 4000, 2500]
})
```

2️⃣ DATA CLEANING

```python
sales_data.dropna(inplace=True)
temperature_data.dropna(inplace=True)
expense_data.dropna(inplace=True)
```

3️⃣ ANALYSIS


```python
total_sales = sales_data["Sales"].sum()
highest_sales_category = sales_data.loc[sales_data["Sales"].idxmax(), "Category"]

avg_temperature = temperature_data["Temperature"].mean()

total_expenses = expense_data["Amount"].sum()
```

4️⃣ VISUALIZATIONS

```python
plt.figure(figsize=(14, 10))

# ---- Bar Chart: Sales by Category ----
plt.subplot(2, 2, 1)
plt.bar(sales_data["Category"], sales_data["Sales"])
plt.title("Sales by Product Category", fontsize=14)
plt.xlabel("Category")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)

# ---- Line Chart: Monthly Temperature ----
plt.subplot(2, 2, 2)
plt.plot(temperature_data["Month"], temperature_data["Temperature"], marker='o')
plt.title("Monthly Temperature Trend", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")

# ---- Pie Chart: Expense Distribution ----
plt.subplot(2, 2, 3)
plt.pie(expense_data["Amount"], labels=expense_data["Category"], autopct="%1.1f%%")
plt.title("Expense Distribution", fontsize=14)

plt.tight_layout()
plt.show()
```

5️⃣ TEXT REPORT OUTPUT

```python
print("\n===== DATA ANALYSIS REPORT =====\n")
print(f"📌 Total Sales: {total_sales}")
print(f"📌 Highest Selling Category: {highest_sales_category}")
print(f"📌 Average Monthly Temperature: {avg_temperature:.2f}°C")
print(f"📌 Total Annual Expenses: {total_expenses}")

print("\n===== INSIGHTS =====")
print("✔ Groceries contribute the highest sales overall.")
print("✔ Temperature gradually increases from winter to summer.")
print("✔ Rent forms the major part of monthly expenses.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sales Data
sales_data = pd.DataFrame({
    "Category": ["Electronics", "Clothing", "Groceries", "Toys", "Furniture"],
    "Sales": [45000, 30000, 50000, 15000, 28000]
})

# Expense Data
expense_data = pd.DataFrame({
    "Category": ["Rent", "Food", "Transport", "Entertainment", "Utilities"],
    "Amount": [15000, 8000, 3000, 4000, 2500]
})
```

preprocessing

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])
data['Month'] = data['Date/Time'].dt.strftime('%b')

monthly_avg_temp = data.groupby('Month')['Temp_C'].mean().reset_index()

# Define the correct order of months
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_avg_temp['Month'] = pd.Categorical(monthly_avg_temp['Month'], categories=month_order, ordered=True)
monthly_avg_temp = monthly_avg_temp.sort_values('Month')

temperature_data = monthly_avg_temp.rename(columns={'Temp_C': 'Temperature'})

print("Monthly average temperature data generated:")
print(temperature_data.head())
```

clean

```python
sales_data.dropna(inplace=True)
temperature_data.dropna(inplace=True)
expense_data.dropna(inplace=True)
data.dropna(inplace=True)
```

Analyze

```python
total_sales = sales_data["Sales"].sum()
highest_sales_category = sales_data.loc[sales_data["Sales"].idxmax(), "Category"]

avg_temperature = temperature_data["Temperature"].mean()

# Identify hottest and coldest months
hottest_month_row = temperature_data.loc[temperature_data["Temperature"].idxmax()]
hottest_month = hottest_month_row["Month"]
coldest_month_row = temperature_data.loc[temperature_data["Temperature"].idxmin()]
coldest_month = coldest_month_row["Month"]

# Calculate average wind speed
avg_wind_speed = data['Wind Speed_km/h'].mean()

# Determine the most frequent weather condition and its counts
weather_counts = data['Weather'].value_counts()
most_frequent_weather = weather_counts.idxmax()

total_expenses = expense_data["Amount"].sum()
```

visualize

```python
plt.figure(figsize=(14, 10))

# ---- Bar Chart: Sales by Category ----
plt.subplot(2, 2, 1)
plt.bar(sales_data["Category"], sales_data["Sales"])
plt.title("Sales by Product Category", fontsize=14)
plt.xlabel("Category")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)

# ---- Line Chart: Monthly Temperature ----
plt.subplot(2, 2, 2)
plt.plot(temperature_data["Month"], temperature_data["Temperature"], marker='o')
plt.title("Monthly Temperature Trend", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")

# ---- Pie Chart: Expense Distribution ----
plt.subplot(2, 2, 3)
plt.pie(expense_data["Amount"][expense_data["Amount"] > 0], labels=expense_data["Category"][expense_data["Amount"] > 0], autopct="%1.1f%%")
plt.title("Expense Distribution", fontsize=14)

# ---- Bar Chart: Frequency of Weather Conditions ----
plt.subplot(2, 2, 4)
plt.bar(weather_counts.index, weather_counts.values)
plt.title("Frequency of Weather Conditions", fontsize=14)
plt.xlabel("Weather Condition")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

HANDS-ON PRACTICE (With starter examples)
1. Time-based message

```python
hour = int(input("Enter hour (0-23): "))

if hour < 12:
    print("Good Morning!")
elif hour < 18:
    print("Good Afternoon!")
else:
    print("Good Evening!")
```

2. Number Guessing (basic version)

```python
secret = 7
guess = int(input("Guess the number: "))

if guess == secret:
    print("Correct!")
else:
    print("Wrong guess! Try again!")
```

4. Function: Area of Rectangle

```python
def area(length, width):
    return length * width

print(area(5, 3))
```

5. Simple Calculator

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose + - * / : ")

if op == "+":
    print(num1 + num2)
elif op == "-":
    print(num1 - num2)
elif op == "*":
    print(num1 * num2)
elif op == "/":
    print(num1 / num2)
else:
    print("Invalid operator")
```

Project:
Analyze a Simple Sales Dataset to find: total sales, best-selling product, and create a basic report

```python
import pandas as pd

sales = pd.read_csv("sales_dataset.csv")
print(sales.head())
```

1. Total Sales

```python
sales["Total"] = sales["Quantity"] * sales["Price"]
total_sales = sales["Total"].sum()

print("Total Sales:", total_sales)
```

2. Best-Selling Product

```python
best_product = (
    sales.groupby("Product")["Quantity"]
    .sum()
    .idxmax()
)

print("Best Selling Product:", best_product)
```

3. Create a Simple Report

```python
report = {
    "Total Sales": total_sales,
    "Best Selling Product": best_product,
    "Total Quantity Sold": sales["Quantity"].sum(),
    "Average Sale Value": sales["Total"].mean()
}

print("\n--- SALES REPORT ---")
for key, value in report.items():
    print(f"{key}: {value}")
```
